```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sbn

df = pd.read_csv(r"C:\Users\Cybrom\Desktop\numpy_assign\Telco-Customer-Churn.csv")
df
```

```python
dimension  = df.shape
print("Number of Columns:",dimension[1])
print("Number of rows:",dimension[0])
df.head()
```

```python
# column name
print("Column Name :",df.columns)
df.dtypes
```

```python
#Tranform TotalCharges into Numerical Field 
df["TotalCharges"] = df["TotalCharges"].replace(" ","0")
df["TotalCharges"] = df["TotalCharges"].astype("float")
df.dtypes
```

```python
#Tranform TotalCharges into Numerical Field (second way)
df.dtypes
df["TotalCharges"] = pd.to_numeric(df["TotalCharges"],errors='coerce')
df.dtypes
```

```python
#Null handling
print("Total Null value column wise\n\n",df.isnull().sum())
print("\nTotal Null value in all columns\n\n",df.isnull().sum().sum())
```

```python
#Replace NaN with 0
df["TotalCharges"] = df["TotalCharges"].replace(np.nan , 0)
print("Total Null value column wise\n\n",df.isnull().sum())
print("\nTotal Null value in all columns\n\n",df.isnull().sum().sum())
```

```python
# summary of our data set
print(df.info())
print("\n")
print(df.describe())
```

```python
# check duplicate
number_duplicate_df = df.duplicated().sum()
print("Number of Duplicate row:",number_duplicate_df)
number_duplicate_pk = df["customerID"].duplicated().sum()
print("Number of Duplicate in PK:",number_duplicate_pk)
```

```python
# Number of churns in our data
plt.figure(figsize = (3,4)) 
ax = sbn.countplot(x = "Churn",data = df)
#print(ax.containers)
ax.bar_label(ax.containers[0])
plt.show()
```

```python
# prepare pie chart
churn_counts = df["Churn"].value_counts()
churn_counts
plt.figure(figsize = (3,3))
plt.pie(
    churn_counts,
    labels = churn_counts.index,
    startangle = 90,
    autopct = "%1.1f%%"
)
plt.title("Percentage Distribution of Churn")
plt.show()
```

```python
# Create a cross-tab (count table) for Churn vs Gender
churn_gender = pd.crosstab(df["gender"], df["Churn"])

# Plot stacked bar chart
churn_gender.plot(
    kind="bar",
    stacked=True,
    figsize=(5,4),
   
)

plt.title("Churn Distribution by Gender")
plt.xlabel("Gender")
plt.ylabel("Number of Customers")
plt.legend(title="Churn")
plt.show()
```
